# Segmentation 
Here we will show how to perform segmentation using classical methods and machine learning. 

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
import seaborn as sns
```

```python
# load the audio file

# load only the middle 30 seconds of the audio file
y, sr = librosa.load('cse.wav', offset=90, duration=30)

# print the shape of the audio file and sampling rate
print("Shape of the audio file:", y.shape, "Sampling rate:", sr)

# play the audio file
ipd.Audio(y, rate=sr)
```

# Method 1: Fixed-Length Segments
We cut the music into segments of 10 seconds each. Then, we calculate the segment's lenght which is defoned as the number of samples in the segment. 

```python
# Let's cut the music into segments of 10 seconds each
segment_duration = 15  # 10 seconds
segment_length = int(segment_duration * sr) # segment length in samples. 

# Split the audio into fixed-length segments
segments = librosa.util.frame(y, frame_length=segment_length, hop_length=segment_length).T

# play each segment 
for i, segment in enumerate(segments):
    print(f"Playing segment {i}")
    ipd.display(ipd.Audio(segment, rate=sr))
```
