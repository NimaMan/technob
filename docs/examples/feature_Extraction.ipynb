# Feature Extraction

```python
import librosa
import matplotlib.pyplot as plt
```

```python
# Load 
audio_file = "cse.wav"
y, sr = librosa.load(audio_file,)
```

```python


# Extract magical features
intensity = librosa.feature.rms(y=y)[0]
pitch, _ = librosa.piptrack(y=y, sr=sr)
pitch = pitch[0]
timbre = librosa.feature.chroma_stft(y=y, sr=sr)

# Let's visualize the magical features
plt.figure(figsize=(12, 6))

# Intensity (loudness) feature
plt.subplot(3, 1, 1)
plt.plot(librosa.times_like(intensity))
plt.title("Intensity - Loudness")
plt.xlabel("Time (seconds)")
plt.ylabel("Intensity")

# Pitch feature
plt.subplot(3, 1, 2)
plt.plot(librosa.times_like(pitch), pitch.T)
plt.title("Pitch - Magical Notes")
plt.xlabel("Time (seconds)")
plt.ylabel("Pitch (Hz)")

# Timbre feature
plt.subplot(3, 1, 3)
librosa.display.specshow(timbre, x_axis='time', y_axis='chroma', cmap='coolwarm')
plt.colorbar()
plt.title("Timbre - Colorful Harmonies")
plt.xlabel("Time (seconds)")
plt.ylabel("Pitch Class")

plt.tight_layout()
plt.show()
```
