# Quantization

```python
import pyrubberband as pyrb
import numpy as np
import math
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
# Load the music file
audio_file = "cse.wav"
y, sr = librosa.load(audio_file)

# Set the desired tempo for quantization (120 BPM)
desired_tempo = 120
# Perform the magical quantization
y_quantized = pyrb.time_stretch(y, sr, 1.0, desired_tempo / sr)


```

```python

# Play the original audio
ipd.Audio(y, rate=sr)
```

```python
# Play the quantized audio
ipd.Audio(y_quantized, rate=sr)
```
