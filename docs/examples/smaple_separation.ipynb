# Sample Separation

```python
import numpy as np
import librosa
import librosa.display as disp
import IPython.display as ipd

import matplotlib.pyplot as plt
```

```python
# load 30 seconds of audio from the middle of the file
y, sr  = librosa.load("cse.WAV", offset=90, duration=30)
```

```python
# Now, let's use the "librosa" library to perform Sample Separation
harmonic, percussive = librosa.effects.hpss(y)

# Play the samples in the notebook
ipd.Audio(harmonic, rate=sr)
```

```python

ipd.Audio(percussive, rate=sr)
```
